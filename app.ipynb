# Automating Amazon Search Term Excel Sheet
### It creates a nice report by extracting orders on different basis:
- Order > 0 from Column: R (7 Day Total Orders (#)), Filter ASIN & Keyword
- Order == 1, ASIN & Keyword
- Order == 2, ASIN & Keyword
- Order == 3, ASIN & Keyword
- Order > 3, ASIN & Keyword

```python
import pandas as pd
import os
import openpyxl as pxl
```

```python
pd.set_option('display.max_columns', 24)
saveTo = 'search_term_pro.xlsx'

xls = pd.ExcelFile('fresh_search_term.xlsx')
df = pd.read_excel(xls, sheet_name='Sponsored Product Search Term R')

# df.head(3)
df
```

```python
(df['7 Day Total Orders (#)'] > 0).head()
```

## Order > 0

```python
orders_greater_than_zero = df[(df['7 Day Total Orders (#)'] > 0)]
orders_greater_than_zero.to_excel('search_term_pro.xlsx', 'Order>0', index=False)


# orders_greater_less_zero = df[(df['7 Day Total Orders (#)'] > 0)]
# # orders_greater_less_zero.to_excel('search_term_pro.xlsx', 'Order<0', index=False)
```

```python
order_greater_than_zero_with_asin = orders_greater_than_zero[orders_greater_than_zero['Customer Search Term'].str.startswith('b0')]
```

```python
# orders_greater_than_zero['Customer Search Term'].str.startswith('b0') == False
```

```python
order_greater_than_zero_with_keyword = orders_greater_than_zero[orders_greater_than_zero['Customer Search Term'].str.startswith('b0') == False]
```

## Order == 1

```python
order_equals_one = orders_greater_than_zero[orders_greater_than_zero['7 Day Total Orders (#)'] == 1]
```

```python
order_equals_one_with_asin = order_equals_one[order_equals_one['Customer Search Term'].str.startswith('b0')]
order_equals_one_with_asin.shape
```

```python
order_equals_one_with_keyword = order_equals_one[order_equals_one['Customer Search Term'].str.startswith('b0') == False]
order_equals_one_with_keyword.shape
```

## Order == 2

```python
orders_equals_two = orders_greater_than_zero[orders_greater_than_zero['7 Day Total Orders (#)'] == 2]
orders_equals_two
```

```python
orders_equals_two_with_asin = orders_equals_two[orders_equals_two['Customer Search Term'].str.startswith('b0')]
orders_equals_two_with_asin.shape
```

```python
orders_equals_two_with_keyword = orders_equals_two[orders_equals_two['Customer Search Term'].str.startswith('b0') == False]
orders_equals_two_with_keyword.shape
```

## Order == 3

```python
orders_equals_three = orders_greater_than_zero[orders_greater_than_zero['7 Day Total Orders (#)'] == 3]
orders_equals_three.shape
```

```python
orders_equals_three_with_asin = orders_equals_three[orders_equals_three['Customer Search Term'].str.startswith('b0')]
orders_equals_three_with_asin
```

## Order > 3

```python
orders_greater_than_three = orders_greater_than_zero[orders_greater_than_zero['7 Day Total Orders (#)'] > 3]
orders_greater_than_three.shape
```

```python
orders_greater_than_three_with_asin = orders_greater_than_three[orders_greater_than_three['Customer Search Term'].str.startswith('b0')]
orders_greater_than_three_with_asin.shape
```

```python
orders_greater_than_three_with_keyword = orders_greater_than_three[orders_greater_than_three['Customer Search Term'].str.startswith('b0') == False]
orders_greater_than_three_with_keyword.shape
```

```python
excel_book = pxl.load_workbook(saveTo)
with pd.ExcelWriter(saveTo, engine='openpyxl') as writer:
    orders_greater_than_zero.to_excel(writer, 'Order>0', index=False)
    order_greater_than_zero_with_asin.to_excel(writer, 'Order>0_with_asin', index=False)
    order_greater_than_zero_with_keyword.to_excel(writer, 'Order>0_with_kw', index=False)
    
    order_equals_one_with_asin.to_excel(writer, 'Order=1_with_asin', index=False)
    order_equals_one_with_keyword.to_excel(writer, 'Order=1_with_kw', index=False)
    
    orders_equals_two_with_asin.to_excel(writer, 'Order=2_with_asin', index=False)
    orders_equals_two_with_keyword.to_excel(writer, 'Order=2_with_kw', index=False)
    
    orders_equals_three_with_asin.to_excel(writer, 'Order=3_with_asin', index=False)
    orders_equals_three_with_keyword.to_excel(writer, 'Order=3_with_kw', index=False)
    
    orders_greater_than_three_with_asin.to_excel(writer, 'Order>3_with_asin', index=False)
    orders_greater_than_three_with_keyword.to_excel(writer, 'Order>3_with_kw', index=False)
```
